```python

import pandas as pd, duckdb
url = "https://raw.githubusercontent.com/tejaskumarac-web/ds-basics/refs/heads/main/data/orders_sample.csv"
df = pd.read_csv(url, parse_dates=["order_date"])
duckdb.sql("SELECT COUNT(*) orders, SUM(amount) revenue FROM df").df()
duckdb.sql("""
  SELECT customer_id, COUNT(*) n_orders, SUM(amount) revenue
  FROM df GROUP BY 1 ORDER BY revenue DESC LIMIT 5
""").df()
```
